```python
# 1. Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\ajasl\OneDrive\Documents\Project\OIBSIP\Task_1\retail_sales_dataset.csv')
```

```python
# 3. Quick structural check
print("First 5 rows of the dataset:")
display(df.head())
```

```python
# 4. Data Cleaning: Convert the 'Date' column to actual datetime objects 
df['Date'] = pd.to_datetime(df['Date'])
```

```python
# 5. Descriptive Statistics
print("\nDescriptive Statistics for Numerical Columns:")
display(df.describe())
```

```python
# 6. Your First Visualization: Sales by Product Category
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Product Category', palette='Set2')
plt.title('Number of Transactions by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()
```

```python
# 7. Time Series Analysis: Total Sales Over Time
plt.figure(figsize=(12, 5))
sales_over_time = df.groupby('Date')['Total Amount'].sum()
plt.plot(sales_over_time.index, sales_over_time.values, color='teal', linewidth=2)
plt.title('Total Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# 8. Customer Analysis: Transactions by Gender
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Gender', palette='Pastel1')
plt.title('Customer Demographics: Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Transactions')
plt.show()
```
