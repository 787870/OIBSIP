```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 1. Load the Airbnb dataset
df = pd.read_csv(r'C:\Users\ajasl\OneDrive\Documents\Project\OIBSIP\Task_3\AB_NYC_2019.csv\AB_NYC_2019.csv')
```

```python
# 2. Inspect the data structure
print("Dataset Shape:", df.shape)
print("\n--- Missing Values Before Cleaning ---")
print(df.isnull().sum()[df.isnull().sum() > 0])
```

```python
# 3. Check for any exact duplicate rows
print("\nTotal duplicate rows:", df.duplicated().sum())
```

```python
# 4. Look at the first few rows
display(df.head())
```

```python
# 1. Fill missing 'name' and 'host_name' with a placeholder
df['name'] = df['name'].fillna('Unknown')
df['host_name'] = df['host_name'].fillna('Unknown')
```

```python
# 2. Fill missing 'reviews_per_month' with 0 
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
```

```python
# 3. Fill missing 'last_review' date with a text placeholder
df['last_review'] = df['last_review'].fillna('Not Reviewed')
```

```python
# 4. The Moment of Truth: Check if we have any missing values left!
print("--- Missing Values AFTER Cleaning ---")
print(df.isnull().sum())
```

```python
# 1. Save the newly cleaned dataset to a new file
df.to_csv('Cleaned_AB_NYC_2019.csv', index=False)
print("✅ Clean dataset saved successfully as 'Cleaned_AB_NYC_2019.csv'!")
```

```python
# 2. Let's make a quick chart to show the data is ready for analysis
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='room_type', palette='Set2')
plt.title('Distribution of Airbnb Room Types in NYC', fontsize=14, fontweight='bold')
plt.xlabel('Room Type', fontsize=12)
plt.ylabel('Total Listings', fontsize=12)
plt.tight_layout()
plt.show()
```
